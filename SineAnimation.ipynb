```julia
t = 1:60 # time in seconds
```

```julia
typeof(t)
```

```julia
ω₀ = 0.1; # frequency in s^-1
```

```julia
w₀ = [sin(x-ω₀*t) for x=x, t=t]
```

```julia
using Plots, Printf
```

```julia
plot(x,w₀[:,1])
```

```julia
anim = @animate for i = 1:length(t)

    titlestr = @sprintf("Sine wave at t = %.1f seconds", t[i])

    w_plot = plot(x, w₀[:,i], linewidth = 2,
                  label = "", xlabel = "x (m)", ylabel = "sin(x-ωt)")

    plot(w_plot, size = (800, 600))
end;
```

```julia
gif(anim, "anim_sine.gif", fps = 15)
```

```julia
ω₁ = 0.12; # frequency in s^-1
```

```julia
w₁ = [sin(x-ω₁*t+0.1) for x=x, t=t];
```

```julia
W = w₀ + w₁;
```

```julia
anim_packet = @animate for i = 1:length(t)

    titlestr = @sprintf("Sine wave at t = %.1f seconds", t[i])

    w_plot = plot(x, W[:,i], linewidth = 2,
                  label = "", xlabel = "x (m)", ylabel = "sin(x-ωt)")

    plot(w_plot, size = (800, 600))
end;
```

```julia
gif(anim_packet, "anim_sine.gif", fps = 15)
```
