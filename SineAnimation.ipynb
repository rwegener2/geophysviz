## Julia basics of creating, plotting and animating a sine wave

```julia
anim_dir = "animations/"
```

```julia
x=1:10
```

```julia
typeof(x)
```

```julia
x = -π:.01π:π
```

```julia
typeof(x)
```

```julia
t = 1:60 # time in seconds
```

```julia
typeof(t)
```

```julia
ω₀ = 0.5; # frequency in s^-1
```

```julia
w₀ = [sin(5*x-ω₀*t) for x=x, t=t]
```

```julia
using Plots, Printf
```

```julia
plot(x,w₀[:,1])
```

```julia
anim = @animate for i = 1:length(t)

    titlestr = @sprintf("Sine wave at t = %.1f seconds", t[i])

    w_plot = plot(x, w₀[:,i], linewidth = 2,
                  label = "", xlabel = "x (m)", ylabel = "sin(x-ωt)")

    plot(w_plot, size = (800, 600))
end;
```

```julia
string(anim_dir,"anim_sine.gif")
```

```julia
gif(anim, string(anim_dir,"anim_sine.gif"), fps = 15)
```

```julia
ω₁ = 0.7; # frequency in s^-1
```

```julia
w₁ = [sin(5.2*x-ω₁*t+0.1) for x=x, t=t];
```

```julia
W = w₀ + w₁;
```

```julia
anim_packet = @animate for i = 1:length(t)

    titlestr = @sprintf("Sine wave at t = %.1f seconds", t[i])

    w_plot = plot(x, W[:,i], linewidth = 2,
                  label = "", xlabel = "x (m)", ylabel = "sin(x-ωt)")

    plot(w_plot, size = (800, 600), xlims=(-π, π), ylims=(-3, 3))
end;
```

```julia
gif(anim_packet, string(anim_dir,"anim_sine_packet.gif"), fps = 15)
```
